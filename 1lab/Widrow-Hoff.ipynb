```python
import numpy as np
import math
```

```python
def act(sum):
    return 1 / (1 + math.exp(-sum))
```

```python
def train(X, Y, W):
  iter = 0
  speed = 1
  sum_perfom = 1
  while sum_perfom / 4 > 0.01:
  #for i in range(2000): # количество повторов
    sum_perfom = 0
    for j in range(0, len(X)): # проход по каждому правилу
      Out = act(np.dot(X[j], W))
      err = Y[j] - Out 
      perfom = err ** 2 / 2
      sum_perfom += perfom * 2
      dw = speed * err * X[j]
      W = W + dw
      #print(dw)
      iter += 1
      print("№"+str(iter), "X:", X[j], "Y:", Y[j], "Out:", Out, "Weights:", W, "Error:", err, "Perf", perfom)
  return W
```

```python
x = [ [0, 0, 1], [1, 0, 1], [0, 1, 1], [1, 1, 1] ] # входы, в данном случае 4 правила, где последний x это сдвиг
y = [ 0, 1, 1, 1 ] # выходы, в данном случае 4 ответа для каждого правила
w = [0, 0, 0] # веса
x = np.array(x)
y = np.array(y)
w = np.array(w)

w = train(x, y, w)
```

```python
from matplotlib import pyplot as plt

def fun(X):
    return (-w[1] * X - w[2]) / w[0]
    
xx = [i for i in range(-1, 3)]
yy = [fun(i) for i in range(-1, 3)]


for i in range(len(x)):
    if y[i] == 1:
        plt.plot(x[i][0], x[i][1], 'y^')
    else:
        plt.plot(x[i][0], x[i][1], 'gs')

plt.plot(xx, yy, 'r')
plt.show()
```
